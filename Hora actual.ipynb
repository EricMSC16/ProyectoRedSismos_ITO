```python
import serial
import pandas as pd
import time
from datetime import datetime

# Configura el puerto serie (ajusta 'COM4' según tu configuración)
ser = serial.Serial('COM4', 115200, timeout=1)  # Cambia 'COM4' según tu configuración
time.sleep(2)  # Esperar a que se establezca la conexión

# Listas para almacenar datos de ambos maestros
data_master1 = []
data_master2 = []

try:
    # Leer datos durante un tiempo determinado (60 segundos)
    start_time = time.time()
    while time.time() - start_time < 180:  # Cambia 60 por el número de segundos que desees
        if ser.in_waiting > 0:  # Si hay datos disponibles para leer
            line = ser.readline().decode('utf-8').strip()  # Leer la línea y decodificar
            print(f"Línea leída: {line}")  # Para depuración

            # Procesar la línea
            if line.startswith("Datos recibidos del Maestro 1:"):
                parts = line.split(':')[1].strip().split(',')
                print(f"Partes Maestro 1: {parts}")  # Para depuración
                if len(parts) == 7:  # Asegurarse de que hay 7 elementos
                    try:
                        # Usar la hora actual si el timestamp está vacío
                        timestamp = parts[0] if parts[0] else datetime.now().isoformat()  # Usar hora actual
                        data_master1.append({
                            'timestamp': timestamp,
                            'accel_x': float(parts[1]),
                            'accel_y': float(parts[2]),
                            'accel_z': float(parts[3]),
                            'gyro_x': float(parts[4]),
                            'gyro_y': float(parts[5]),
                            'gyro_z': float(parts[6]),
                        })
                    except ValueError as e:
                        print(f"Error al convertir datos del Maestro 1: {e}")
            elif line.startswith("Datos recibidos del Maestro 2:"):
                parts = line.split(':')[1].strip().split(',')
                print(f"Partes Maestro 2: {parts}")  # Para depuración
                if len(parts) == 7:  # Asegurarse de que hay 7 elementos
                    try:
                        # Usar la hora actual si el timestamp está vacío
                        timestamp = parts[0] if parts[0] else datetime.now().isoformat()  # Usar hora actual
                        data_master2.append({
                            'timestamp': timestamp,
                            'accel_x': float(parts[1]),
                            'accel_y': float(parts[2]),
                            'accel_z': float(parts[3]),
                            'gyro_x': float(parts[4]),
                            'gyro_y': float(parts[5]),
                            'gyro_z': float(parts[6]),
                        })
                    except ValueError as e:
                        print(f"Error al convertir datos del Maestro 2: {e}")

except KeyboardInterrupt:
    print("Finalizando la lectura de datos.")

finally:
    # Cerrar el puerto serie
    ser.close()

# Procesar los datos solo si hay datos leídos
if data_master1:
    df_master1 = pd.DataFrame(data_master1)
    df_master1.to_csv('datos_transmisorp1801.csv', index=False)
    print("Datos del Maestro 1 guardados en datos_maestro1.csv.")
else:
    print("No se recibieron datos del Maestro 1.")

if data_master2:
    df_master2 = pd.DataFrame(data_master2)
    df_master2.to_csv('datos_transmisorp1802.csv', index=False)
    print("Datos del Maestro 2 guardados en datos_maestro2.csv.")
else:
    print("No se recibieron datos del Maestro 2.")
```
